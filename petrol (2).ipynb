```python
df['Date']= pd.to_datetime(df['Date'])


plt.figure(figsize=(10,6))
plt.plot(df['Date'], df['Delhi'], label='Petroleum Price Delhi',color='r')
plt.plot(df['Date'],df['Kolkata'], label='Petroleum Price Kolkata',color='b')
plt.plot(df['Date'],df['Mumbai'], label='Petroleum Price Mumbai',color='g')
plt.plot(df['Date'],df['Chennai'], label='Petroleum Price Chennai',color='#FF69B4')

plt.title('Petroleum Stock Price Over Time',color='g')
plt.xlabel('Date',color='g')
plt.ylabel('Price',color='g')
plt.grid(True)
plt.legend()
plt.show()
```

### This graph provides a visual insight into the trends and dynamics of petroleum prices over time, helping to identify how prices have evolved and trends such as seasonal fluctuations or market volatility.

```python
# Add a rolling average with a 30-day window
df['Rolling_Avg'] = df['Delhi'].rolling(window=30).mean()
# Plot rolling average along with the original data
plt.figure(figsize = (10, 6))
plt.plot(df['Date'], df['Delhi'], label = "Original Price Delhi", alpha = 0.5)
plt.plot(df['Date'], df['Kolkata'], label = "Original Price Kolkata", alpha = 0.5)
plt.plot(df['Date'], df['Mumbai'], label = "Original Price Mumbai", alpha = 0.5)
plt.plot(df['Date'], df['Chennai'], label = "Original Price Chennai", alpha = 0.5)
plt.plot(df['Date'], df['Rolling_Avg'], label = "30 Days Rolling Avg" , color = 'red')
plt.title('Petroleum Stock Price and Rolling Average')
plt.xlabel('Year')
plt.ylabel('Price')
plt.grid(True)
plt.legend()
plt.show()
```

### This graph helps analyze petroleum price data, revealing short-term price movements and more stable long-term trends.  It is particularly useful for understanding whether price spikes are just temporary or part of a wider trend.
